```python
num_df = df.select_dtypes(include=["int64", "float64"])

# seuil : au moins 60% de valeurs non-nulles
threshold = 0.6  

cols_ok = num_df.columns[num_df.notna().mean() > threshold]

num_df_filtered = num_df[cols_ok]

corr = num_df_filtered.corr()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Corrélation — colonnes numériques conservées (>=60% valeurs)")
plt.show()
```

```python
df["building_age"] = 2025 - df["built_year"]

# Tranches d'âge
df["age_group"] = pd.cut(
    df["building_age"],
    bins=[0, 20, 50, 150],
    labels=["new", "mid", "old"],
    include_lowest=True
)

# Indique si orientation connue
df["has_orientation"] = df[["is_orientation_north", "is_orientation_south", "is_orientation_east"]].notna().any(axis=1)
```

```python
cols_to_drop = [
    'latitude', 'longitude',      # toujours 0
    'portal', 'door',             # toujours 0
    'n_floors', 'sq_mt_allotment',# > 90% NA
    'Unnamed: 0',                 # index inutile
    'title', 'subtitle',          # pas informatif pour un modèle tabulaire
]
df = df.drop(columns=cols_to_drop, errors='ignore')
```

```python
df['sq_mt_built'] = df['sq_mt_built'].fillna(df['sq_mt_built'].median())
df['sq_mt_useful'] = df['sq_mt_useful'].fillna(df['sq_mt_useful'].median())
df['n_bathrooms'] = df['n_bathrooms'].fillna(df['n_bathrooms'].median())
df['built_year'] = df['built_year'].fillna(df['built_year'].median())
df['building_age'] = df['building_age'].fillna(df['building_age'].median())
```

```python
bool_cols = [
    'is_furnished', 'are_pets_allowed', 'is_kitchen_equipped',
    'has_private_parking', 'has_public_parking',
    'is_orientation_north', 'is_orientation_south', 'is_orientation_east',
]

for col in bool_cols:
    df[col] = df[col].astype('bool', errors='ignore')
    df[col] = df[col].fillna(False)

df["has_orientation"] = df["has_orientation"].fillna(False)
```

```python
cat_cols = df.select_dtypes(include='object').columns
df[cat_cols] = df[cat_cols].fillna("Unknown")
```

```python

num_df = df.select_dtypes(include=["int64", "float64"])

corr_matrix = num_df.corr(method="pearson")

plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, cmap="viridis", linewidths=.5, center=0)
plt.title("Matrice de corrélation (Pearson) — Variables numériques")
plt.show()
```

```python
from scipy.stats import pointbiserialr

bool_cols = df.select_dtypes(include='bool').columns
target = df["buy_price"]

pb_corr = {}

for col in bool_cols:
    corr, pval = pointbiserialr(df[col], target)
    pb_corr[col] = corr

pd.DataFrame.from_dict(pb_corr, orient="index", columns=["Corr_point_biserial"]).sort_values("Corr_point_biserial", ascending=False)
```

```python
results = {}

for col in bool_cols:
    mean_true = df[df[col] == True]["buy_price"].mean()
    mean_false = df[df[col] == False]["buy_price"].mean()
    results[col] = mean_true - mean_false

pd.DataFrame.from_dict(results, orient="index", columns=["Impact_sur_le_prix"]).sort_values("Impact_sur_le_prix", ascending=False)
```

```python
# Colonnes booléennes mal typées
bool_cols = [
    "is_floor_under",
    "is_new_development",
    "has_central_heating",
    "has_individual_heating",
    "has_ac",
    "has_fitted_wardrobes",
    "has_lift",
    "is_exterior",
    "has_garden",
    "has_pool",
    "has_terrace",
    "has_balcony",
    "has_storage_room",
    "is_accessible",
    "has_green_zones",
    "is_parking_included_in_price",
    "is_orientation_west"
]

def to_bool(x):
    if isinstance(x, str):
        x = x.lower().strip()
        if x in ("true", "yes", "1", "si"):
            return True
        if x in ("false", "no", "0", "non"):
            return False
    return pd.NA

for col in bool_cols:
    df[col] = df[col].apply(to_bool).astype("boolean")
```

```python
cat_cols = [
    "raw_address",
    "street_name",
    "street_number",
    "floor",
    "neighborhood_id",
    "operation",
    "house_type_id",
    "energy_certificate"
]

for col in cat_cols:
    df[col] = df[col].astype("category")
```
